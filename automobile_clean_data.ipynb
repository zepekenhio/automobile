```python
#importer les bibliothèques
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('automobile.csv')
```

```python
#vérifier le type de données
df.dtypes
```

```python
#afficher les 5 premières lignes du jeu de données
df.head(5)
```

```python
del df['Unnamed: 0']
```

```python
df.shape
```

```python
# nettoyer les données en éliminant les valeurs NaN et 0
df.isna().sum()
```

```python
#Afficher le détail de toutes les voitures de marque Toyota
df.loc[df.company == "toyota"]
```

```python
#Afficher le nom de la compagnie, le nombre de chevaux ainsi que le prix de la voiture la plus chère
df[df['price']==df['price'].max()].get(['company','horsepower','price'])
```

```python
#Afficher le nombre de voitures par compagnie
df['company'].value_counts()
```

```python
#afficher pour chaque compagnie le prix de la voiture le plus élevé
df.groupby('company').max().get(['price'])
```

```python
#afficher la moyenne de kilométrage par compagnie
df.groupby(by='company').mean('average-mileage').get(['average-mileage'])
```

```python
#trier toutes les voitures par prix
df.sort_values('price', ascending=False).get(['company','price'])
```
